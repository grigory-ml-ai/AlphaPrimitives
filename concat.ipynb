```python
import pandas as pd
import polars as pl
from pathlib import Path
```

```python
filepath = Path(r"C:\Users\LianLi\Desktop\trading\dataset\futures\BNBUSDC-2024-01-2025-12.parquet")
features = pl.scan_parquet(filepath).collect_schema().names()
features
```

```python
def download_polars(path: Path, feature_name: str) -> pl.LazyFrame:
    return pl.scan_parquet(path).select('close_time', feature_name)
```

```python
download_polars(Path(r"C:\Users\LianLi\Desktop\trading\dataset\futures\BTCUSDT-2020-01-2025-12.parquet"), 'close').collect()
```

```python
df = pl.concat(
    [
        download_polars(path, 'close')
        for path in Path(r'C:\Users\LianLi\Desktop\trading\dataset\futures').glob('*.parquet')
    ], how='vertical'
).sink_parquet(r'C:\Users\LianLi\Desktop\trading\futures_close.parquet')
```
